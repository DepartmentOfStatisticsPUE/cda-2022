```python
%%R
install.packages("Rlab")
library(Rlab)

y <- rbern(10000, 0.7)

lambd <- 0.5 * y + 0.5
```

```python
%%R
pois <- rpois(10000, lambd)
plot(pois, type='l')
```

```python
%%R
hist(pois)
sr = mean(pois)
sr
```
